What is Numpy Array?

It is an Array class called ndarray where N represents N dimensional array and it can be 1, 2, 3 or more.

Characteristics of Numpy array:

1. All the elements are of same datatype

2. Once numpy arary is created total size of the array cannot be changed

3. In a 2D array each row of have the same no of columns

```python
import numpy as np
```

```python
#1D array

a1 = np.array([1,2,3,4,5,6])
print(a1)
```

```python
#2D Array

a2 = np.array([[1,2,3,4],[5,6,7,8],[2,3,4,8]])
print(a2)
```

```python
print(a1[3])
```

```python
print(a2[1,3])
```

```python
#1-D array

print(np.ndim(a1))
print(np.shape(a1))
print(np.size(a1))
print(a1.dtype)
```

```python
#2-D array

print(np.ndim(a2))
print(np.shape(a2))
print(np.size(a2))
print(a2.dtype)
```
