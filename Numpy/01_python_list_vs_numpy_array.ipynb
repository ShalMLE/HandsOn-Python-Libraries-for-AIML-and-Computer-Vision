```python
#Import libraries
import numpy as np
import time
```

```python
# Numpy = Numerical Python
# Backbone of ML, DL, CV and DS

#Vector Addition
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

c = a + b
print(c)

#Scalar Multiplication
weights = np.array([0.2, 0.5, 0.1])
learning_rate = 0.01

updated_weights = weights * learning_rate
print(updated_weights)

#Dot Product
features = np.array([1.5, 2.0, 3.0])
weights = np.array([0.4, 0.3, 0.2])

prediction = np.dot(features, weights)
print(prediction)

#Matrix Multiplication
X = np.array([[1, 2],
              [3, 4],
              [5, 6]])

W = np.array([[0.1],
              [0.2]])

output = np.dot(X, W)
print(output)
```

```python
#Python List Example
lis = [1,2,3,4,5]
print(lis)
print(type(lis))
```

```python
#Numpy Array Example
arr = np.array(lis)
print(arr)
print(type(arr))
```

```python
#Key Difference between Python List and Numpy Array - Element-wise Operation
lis_add = lis * 2
print(lis_add)

arr_add = arr * 2
print(arr_add)
```

```python
#Performance Comparision
large_list = list(range(1_000_000))
large_array = np.array(large_list)

start = time.time()
[x * 2 for x in large_list]
print("List time:", time.time() - start)

start = time.time()
large_array * 2
print("NumPy time:", time.time() - start)
```

```python
#Summary
# - Python List are flexible but slow for math
# - Numpy arrays are fast and support vectorized operations
# - AI/ML relies heavily on numerical computation
```
